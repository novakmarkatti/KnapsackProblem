```python
import json
import time
from cleancode import *
import matplotlib.pyplot as plt
```

```python
def knapsack(approach, capacity, weights, values, n):
    if approach == "knapsackBruteForce":
        startTime = time.time()
        maxValue, maxWeight, selectedItems = knapsackBruteForce(capacity, weights, values, n)
        endTime = time.time() - startTime
    elif approach == "knapsackDivideAndConquer":
        startTime = time.time()
        maxValue, maxWeight, selectedItems = knapsackDivideAndConquer(capacity, weights, values, n)
        endTime = time.time() - startTime
    elif approach == "knapsackMemoization":
        startTime = time.time()
        maxValue, maxWeight, selectedItems = knapsackMemoization(capacity, weights, values, n)
        endTime = time.time() - startTime
    elif approach == "knapsackDP":
        startTime = time.time()
        maxValue, maxWeight, selectedItems = knapsackDP(capacity, weights, values, n)
        endTime = time.time() - startTime
    elif approach == "knapsackGreedy":
        startTime = time.time()
        maxValue, maxWeight, selectedItems = knapsackGreedy(capacity, weights, values, n)
        endTime = time.time() - startTime
    elif approach == "knapsackBacktrackGreedy":
        startTime = time.time()
        maxValue, maxWeight, selectedItems = knapsackBacktrackGreedy(capacity, weights, values, n)
        endTime = time.time() - startTime
    elif approach == "knapsackBacktrack":
        startTime = time.time()
        maxValue, maxWeight, selectedItems = knapsackBacktrack(capacity, weights, values, n)
        endTime = time.time() - startTime
    elif approach == "knapsackBranchAndBound":
        startTime = time.time()
        maxValue, maxWeight, selectedItems = knapsackBranchAndBound(capacity, weights, values, n)
        endTime = time.time() - startTime
    elif approach == "knapsackGenetic":
        startTime = time.time()
        maxValue, maxWeight, selectedItems = knapsackGenetic(capacity, weights, values, n)
        endTime = time.time() - startTime
    elif approach == "knapsackRandomized":
        startTime = time.time()
        maxValue, maxWeight, selectedItems = knapsackRandomized(capacity, weights, values, n)
        endTime = time.time() - startTime                          
    selectedItems.sort(reverse=True)    
    print(approach, "n:", n, "capacity:", capacity, "endTime:", endTime, "maxValue:",maxValue, "maxWeight:",maxWeight, selectedItems)
    return (endTime, maxValue, maxWeight, selectedItems)
```

```python
inputFile = open('../tests/test14.json')
jsonArray = json.load(inputFile)

plt.title("Futs id vizsglata a kapacits nvelse ltal") 
plt.xlabel("Kapacits") 
plt.ylabel("Futs id (log skla)")
plt.grid()
plt.gca().set_xscale('linear')
plt.gca().set_yscale('linear')

approaches = [ "knapsackBruteForce", "knapsackDivideAndConquer", "knapsackMemoization","knapsackDP", "knapsackGreedy", "knapsackBacktrackGreedy", "knapsackBacktrack", "knapsackBranchAndBound", "knapsackGenetic", "knapsackRandomized"]  

for i in range(0, len(approaches)):
               
    xpoints, ypoints = [], []           
    for test in jsonArray:
        capacity = test['capacity']
        items    = test['numberOfItems']
        weights  = test['weights']
        values   = test['values']
        
        endTime, maxValue, maxWeight, selectedItems = knapsack(approaches[i], capacity, weights, values, items) 
        xpoints.append(capacity)
        ypoints.append(endTime if endTime > 0 else 0.0001)   
    plt.plot(xpoints, ypoints, label = approaches[i] , marker = 'o')
    
plt.xticks(xpoints)
plt.yscale('log') 
plt.ylim(bottom=0.00001)  
plt.legend(title="Stratgik", loc='center left', bbox_to_anchor=(1, 0.5))   
plt.show()    
```
